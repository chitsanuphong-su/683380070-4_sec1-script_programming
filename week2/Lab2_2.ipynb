```python
age = input("Enter your age: ")
age = int(age)

is_like_action = input("Do you like action movies? (y/n): ")

if age >= 18 and is_like_action == 'y':
  print("Recommended for ages 18+ and you can watch action movies.")
elif age >= 18 and is_like_action == 'n':
  print("Recommended for ages 18+.")
elif age >= 13 and is_like_action == 'y':
  print("Recommended for ages 13+(13-17) and you should watch action movies with your parents.")
elif age >= 13 and is_like_action == 'n':
  print("Recommended for ages 13+(13-17).")
elif age < 13 and is_like_action == 'y':
  print("Recommended for Safe content appropriate for all family members.\nYou can't watch action movies.")
else:
  print("Recommended for Safe content appropriate for all family members.")
```
