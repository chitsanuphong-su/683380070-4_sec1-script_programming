```python
txt = input("Enter your text: ")

num = int(txt)

if num > 0:
  print(f"{num} is Positive.")
elif num < 0:
  print(f"{num} is Negative.")
else:
  print(f"{num} is Zero.")

if num % 2 == 0:
  print(f"{num} is Even.")
else:
  print(f"{num} is Odd.")
```

```python
txt = input("Enter your text: ")

num = int(txt)

if num > 0:
  if num % 2 == 0:
    print(f"{num} is Positive and Even.")
  else:
    print(f"{num} is Positive and Odd.")
elif num < 0:
  if num % 2 == 0:
    print(f"{num} is Negative and Even.")
  else:
    print(f"{num} is Negative and Odd.")
else:
  print(f"{num} is Zero.")
```
