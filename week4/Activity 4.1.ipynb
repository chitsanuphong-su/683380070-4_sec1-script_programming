## 1.ข้อมูลอุณหภูมิประจำวันที่มีการบันทึกไว้ตลอดหนึ่งสัปดาห์

* **ควรใช้:** **`List`**
* **เหตุผล:** ในระหว่างสัปดาห์เราต้องทยอยเพิ่มข้อมูลทีละวัน (`.append()`) หรือหากคีย์อุณหภูมิผิดพลาดก็สามารถกลับมาแก้ไขค่าในวันนั้นๆ ได้
* **ตัวอย่าง Code:**

```python
# เริ่มต้นบันทึกอุณหภูมิทีละวัน
weekly_temp = [32.5, 33.0, 31.8]

# บันทึกวันถัดไปเพิ่มเข้ามา
weekly_temp.append(34.2)

# หากแก้ข้อมูลอุณหภูมิวันแรกจาก 32.5 เป็น 31.5
weekly_temp[0] = 31.5

print("อุณหภูมิสะสม:", weekly_temp)
```

## 2.ค่าสี RGB ของเฉดสีเฉพาะเฉดหนึ่ง (เช่น สีแดงคือ (255, 0, 0))

* **ควรใช้:** **`Tuple`**
* **เหตุผล:** ค่า RGB ของเฉดสีนั้นๆ เป็นค่าคงที่ที่ไม่ควรถูกเปลี่ยนแปลงเด็ดขาด หากตัวเลขเปลี่ยน สีก็จะเปลี่ยนเป็นสีอื่นทันที การใช้ Tuple ช่วยล็อกค่าไว้เพื่อความปลอดภัย
* **ตัวอย่าง Code:**

```python
# กำหนดค่าสีแบบคงที่ไว้
COLOR_RED = (255, 0, 0)
COLOR_SKY_BLUE = (135, 206, 235)

# หากพยายามจะแอบเปลี่ยนค่าสี โปรแกรมจะแจ้ง Error ทันที
# COLOR_RED[0] = 200  TypeError: 'tuple' object does not support item assignment

print("ค่าสีแดง:", COLOR_RED)
```

## 3.รายการสินค้าในรถเข็นชอปปิงที่สามารถกดเพิ่มหรือลบออกได้

* **ควรใช้:** **`List`**
* **เหตุผล:** พฤติกรรมของรถเข็นชอปปิงต้องยืดหยุ่น ลูกค้าสามารถกดหยิบสินค้าใส่เพิ่ม (`.append()`) หรือลบสินค้าที่ไม่ต้องการออก (`.remove()`) ได้ตลอดเวลา
* **ตัวอย่าง Code:**

```python
shopping_cart = ["นมสด", "ขนมปัง"]

# เพิ่มสินค้าเข้า รถเข็น
shopping_cart.append("ไข่ไก่")

# ลบสินค้าออกจากรถเข็น
shopping_cart.remove("นมสด")

print("สินค้าในรถเข็น:", shopping_cart)
```

## 4.รายชื่อวันในหนึ่งสัปดาห์เรียงตามลำดับ (จันทร์ - อาทิตย์)

* **ควรใช้:** **`Tuple`**
* **เหตุผล:** โครงสร้างของวันในหนึ่งสัปดาห์มี 7 วันตายตัว ไม่มีการเพิ่มวันใหม่ ลบวันออก หรือสลับตำแหน่งกัน การใช้ Tuple จะช่วยประหยัดหน่วยความจำและทำให้โค้ดทำงานได้ไวยิ่งขึ้น
* **ตัวอย่าง Code:**

```python
DAYS_OF_WEEK = (
    "Monday",
    "Tuesday",
    "Wednesday",
    "Thursday",
    "Friday",
    "Saturday",
    "Sunday",
)

# ดึงชื่อวันแรกของสัปดาห์มาใช้งาน
print("วันแรกของสัปดาห์คือ:", DAYS_OF_WEEK[0])
```

## 5.คะแนนเก็บของนักเรียนจากการส่งงานชุดต่างๆ ซึ่งอาจมีการอัปเดตหรือแก้ไขคะแนนในภายหลังได้

* **ควรใช้:** **`List`**
* **เหตุผล:** โจทย์ระบุชัดเจนว่าอาจต้องมีการ "อัปเดตหรือแก้ไข" คะแนนในภายหลัง หรืออาจมีการส่งงานชิ้นใหม่เพิ่มเข้ามา รายชื่อคะแนนจึงจำเป็นต้องแก้ไขข้อมูลเดิมได้
* **ตัวอย่าง Code:**

```python
student_scores = [10, 8, 7]  # คะแนนงานชิ้นที่ 1, 2, 3

# อาจารย์ให้โอกาสแก้ตัว แก้ไขคะแนนงานชิ้นที่ 2 จาก 8 เป็น 10
student_scores[1] = 10

# เพิ่มคะแนนงานชิ้นที่ 4 เข้าไป
student_scores.append(9)

print("คะแนนปัจจุบัน:", student_scores)
```

## 6.พิกัด GPS ของสถานที่สำคัญหรือจุดแลนด์มาร์กที่ตั้งอยู่กับที่

* **ควรใช้:** **`Tuple`**
* **เหตุผล:** สถานที่สำคัญหรือแลนด์มาร์กไม่มีการเคลื่อนที่ พิกัด ละติจูด และ ลองจิจูด (Latitude, Longitude) จึงเป็นค่าคงที่เสมอ การใช้ Tuple ป้องกันไม่ให้พิกัดถูกแก้ไขโดยไม่ได้ตั้งใจ
* **ตัวอย่าง Code:**

```python
# พิกัด GPS ของอนุสาวรีย์ชัยสมรภูมิ (Latitude, Longitude)
VICTORY_MONUMENT_GPS = (13.7649, 100.5383)

print(f"ละติจูด: {VICTORY_MONUMENT_GPS[0]}")
print(f"ลองจิจูด: {VICTORY_MONUMENT_GPS[1]}")
```
