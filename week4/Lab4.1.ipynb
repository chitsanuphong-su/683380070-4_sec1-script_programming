# Lab 4.1: Student Management System (List Operations)
วัตถุประสงค์: ให้นักเรียนสร้างโปรแกรมง่ายๆ เพื่อจัดการรายชื่อนักเรียน เพื่อฝึกฝนและทำความเข้าใจเกี่ยวกับการสั่งงาน List (รายการข้อมูล) ในรูปแบบต่างๆ
สิ่งที่ต้องทำ

---



```python
# 1. Create list of student name
student_names = []

# 2. Input student name in to list
n_student = int(input("Number of student: "))
for i in range(n_student):
  name = input(f"student[{i+1}]: ")
  student_names.append(name)
print(student_names)

# 3. Search student in list
print()
search_student = input("Search student: ")
if search_student in student_names:
  print(f"Found Student! Student {search_student} is in index[{student_names.index(search_student)}].")
else:
  print("Not found student!")

# 4. Remove student in list
print()
wanna_remove = input("Do you want to remove student? (y/n): ")
if wanna_remove == "y":
  remove_student = input("Remove student: ")
  if remove_student in student_names:
    student_names.remove(remove_student)
  else:
    print("Not found student!")

print(student_names)

# 5. Sort Student name (ก-ฮ / A-Z)
print()
student_names.sort()
print(student_names)

# 6. Count studnet in list
print()
print(f"len of student_name is {len(student_names)}")
```

# โจทย์ท้าทาย (Challenge):
ลองสร้างระบบเมนูสั่งการแบบง่ายๆ โดยใช้การวนลูป while (จากบทเรียนสัปดาห์ที่ 3) เพื่อให้ผู้ใช้สามารถกดเลือกได้ว่าจะ เพิ่ม, ลบ, ค้นหา, เรียงลำดับ หรือกดออกจากโปรแกรม

---



```python
student_names = []

while True:

  print("1. เพิ่ม")
  print("2. ลบ")
  print("3. ค้นหา")
  print("4. เรียงลำดับ")
  print("5. ออกจากโปรแกรม\n")

  choice = int(input("Choose your choice(1-5): "))

  if choice == 1:
    n_student = int(input("Number of student: "))
    for i in range(n_student):
      name = input(f"student[{i+1}]: ")
      student_names.append(name)
    print(student_names)
    print()

  elif choice == 2:
    wanna_remove = input("Do you want to remove student? (y/n): ")
    if wanna_remove == "y":
      remove_student = input("Remove student: ")
      if remove_student in student_names:
        student_names.remove(remove_student)
    else:
      print("Not found student!")
    print(student_names)
    print()

  elif choice == 3:
    search_student = input("Search student: ")
    if search_student in student_names:
      print(f"Found Student! Student {search_student} is in index[{student_names.index(search_student)}].")
    else:
      print("Not found student!")
    print()

  elif choice == 4:
    print()
    student_names.sort()
    print(student_names)
    print()

  elif choice == 5:
    print("Exit\n")
    break

  else:
    print("Invalid choice!\n")
```
