# Activity 4.2: Interactive To-Do List (using List Operations)
สิ่งที่อยากให้ทำ : ให้นักเรียนพัฒนาโปรแกรมจัดการรายการสิ่งที่ต้องทำ (To-Do List) แบบข้อความ (Text-based) อย่างง่าย โดยประยุกต์ใช้ความรู้เรื่องการจัดการ List ที่ได้เรียนมา

---



```python
todo_list = []

while True:

  print("1. เพิ่มรายการงาน (Add a task)")
  print("2. ดูรายการงานทั้งหมด (View tasks)")
  print("3. ทำเครื่องหมายว่างานเสร็จสิ้นแล้ว (ลบงานออก โดยเลือกจากตำแหน่ง Index หรือตามชื่อคำสั่ง)")
  print("4. ออกจากโปรแกรม (Exit)\n")

  choice = int(input("Choose your choice: "))

  if choice == 1:
    add_task = input("Enter your task: ")
    todo_list.append(add_task)
    print()

  elif choice == 2:
    print(todo_list)
    print()

  elif choice == 3:
    remove_task = input("Remove your task: ")
    if remove_task in todo_list:
      todo_list.remove(remove_task)
      print(f"Remove {remove_task} complete!")
    else:
      print("No task in todo list")
    print()

  elif choice == 4:
    print("Exit\n")
    break

  else:
    print("Invalid choice!\n")
```
