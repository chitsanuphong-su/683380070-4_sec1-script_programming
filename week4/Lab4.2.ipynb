# Lab 4.2: Geometric Calculations (Tuples & Immutability)
วัตถุประสงค์: ให้นักเรียนใช้ Tuple ในการแทนค่าพิกัด 2 มิติ (2D points) และคำนวณหาระยะห่างระหว่างจุด เพื่อให้เห็นภาพชัดเจนเกี่ยวกับคุณสมบัติ Immutability (การไม่สามารถเปลี่ยนแปลงค่าได้ของ Tuple)


---



```python
import math

point1 = (3, 4)
point2 = (6, 8)

print(f"Point1: x = {point1[0]}, y = {point1[1]}")
print(f"Point2: x = {point2[0]}, y = {point2[1]}")

#point1[0] = 5

distance = math.sqrt((point2[0]-point1[0])**2 + (point2[1]-point1[1])**2)
print(f"Distance = {distance}")
```

# โจทย์ท้าทาย (Challenge):
ลองเขียนโค้ดให้รับโดยใช้คำสั่ง input แทนที่จะกำหนดค่าในโปรแกรม แล้วนำมาคำนวณหาระยะห่าง อย่าลืมแปลงชนิดข้อมูล (Type Conversion) ที่ได้มาจากคำสั่ง input() ให้เรียบร้อยก่อนนำไปคำนวณ (เช่น แปลงเป็น integer หรือ float)

---



```python
import math

x1, y1 = input("x1 y1: ").split()
x2, y2 = input("x2 y2: ").split()

point1 = (float(x1), float(y1))
point2 = (float(x2), float(y2))

print(f"\nPoint1: x = {point1[0]}, y = {point1[1]}")
print(f"Point2: x = {point2[0]}, y = {point2[1]}")

distance = math.sqrt((point2[0]-point1[0])**2 + (point2[1]-point1[1])**2)
print(f"\nDistance = {distance}")
```
