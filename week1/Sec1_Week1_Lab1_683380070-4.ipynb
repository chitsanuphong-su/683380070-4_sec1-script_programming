# Lab 1.2: Working with Variables and Basic Data Type

### 2. Declare and print variables

```python
Age = 19
GPA = 3.78
Favorite_color = "green"
Student_status = True

print("Age =", Age, type(Age))
print("GPA =", GPA, type(GPA))
print("Favorite_color =", Favorite_color, type(Favorite_color))
print("Student_status =", Student_status, type(Student_status))
```

### 3. Perform simple arithmetic

```python
num1 = 5
num2 = 10

sum = num1 + num2
diff = num2 - num1
product = num1 * num2
division = num2 / num1

print("Sum:", sum)
print("Difference:", diff)
print("Product:", product)
print("Division:", division)
```

### 4. String concatenation and f-strings

```python
first_name = "Chitsanuphong"
last_name = "Su"
full_name = first_name + " " + last_name

print(f"Hello, {full_name}!")
```

### 5. User input and type conversion

```python
number = input("What is your favorite number? ")
print(type(number))

number = int(number)
result = number + 5

print("result is", result)
```
