# Lab 3.1: โปรแกรมสร้างสูตรคูณ (Multiplication Table Generator)

### วัตถุประสงค์: ให้นักศึกษาใช้ลูป for (และอาจใช้ลูปซ้อนลูป) เพื่อสร้างแม่สูตรคูณสำหรับตัวเลขที่กำหนด หรือสร้างตารางสูตรคูณตามช่วงตัวเลขที่ระบุ


---



## Part 1: ตารางสูตรคูณแม่เดียว (Single Number Table):

```python
# Part 1: ตารางสูตรคูณแม่เดียว (Single Number Table):

num = input("Enter your number: ")
num = int(num)

print("\n=== Single Number Table ===")

for i in range(1, 13):
  result = num * i
  print(f"{num} x {i} = {result}")
```

## Part 2: ตารางสูตรคูณแบบเต็ม (Full Multiplication Grid) - โจทย์ท้าทาย/ส่วนขยาย:

```python
# Part 2: ตารางสูตรคูณแบบเต็ม (Full Multiplication Grid) - โจทย์ท้าทาย/ส่วนขยาย:

for i in range(1, 13):
  print(i, end="\t")
  for j in range(1, 13):
    result = i * j
    print(result, end="\t")
  print()
```
