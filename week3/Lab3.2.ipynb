### Part 1: โปรแกรมตัวนับถอยหลัง (Countdown Timer):

```python
start_time = int(input("Enter the countdown start time: "))

while start_time > 0:
    print(start_time)
    start_time -= 1

print("Blast off!")
```

### Part 2: เกมทายตัวเลขอย่างง่าย (Simple Guessing Game):

```python
secret_number = 29

while True:
  num = int(input("Guess the number: "))

  if num > secret_number:
    print(f"{num} is Too high! Try again.")
  elif num < secret_number:
    print(f"{num} is Too low! Try agin.")
  else:
    print("Congratulations! You guessed it!")
    break
```
