# กิจกรรมที่ 3.1: จับผิดและแก้ไขโค้ดลูปพัง (Debugging Infinite Loops & Logic Errors)


---



**รูปแบบการทำกิจกรรม:** ทำเดี่ยวหรือจับคู่ จากนั้นนำมาอภิปรายและเฉลยร่วมกันในห้องเรียน

**จุดประสงค์:** เพื่อฝึกทักษะการหาข้อผิดพลาด (Debugging) อย่างมีวิจารณญาณ และทำความเข้าใจข้อผิดพลาดที่มักจะเกิดขึ้นบ่อยๆ เมื่อใช้ลูป

**คำสั่งสำหรับนักศึกษา:**
อาจารย์จะแจกโค้ดที่มีข้อผิดพลาด (Buggy Snippets) ให้ 3 แบบ ให้นักศึกษาวิเคราะห์ว่า **"ทำไมโค้ดถึงทำงานไม่ถูกต้อง"** และ **"ต้องแก้โค้ดอย่างไรให้ทำงานได้ตามเป้าหมาย"**

**ตัวอย่างโค้ดที่มีปัญหา (ให้นักศึกษาแก้):**

**ปัญหาที่ 1: ลูปนรก (Infinite Loop)**
*ปัญหา:* โปรแกรมนี้จะรันเลข 0 ไปเรื่อยๆ ไม่ยอมหยุด เพราะลืมเพิ่มค่าให้กับตัวแปรควบคุม

```python
counter = 0
while counter < 5:
    print(counter)
    # ต้องเติมโค้ดอะไรตรงนี้ เพื่อให้ลูปหยุดได้?

```

**ปัญหาที่ 2: ใช้ `range()` ผิดพลาด (Incorrect Range)**
*ปัญหา:* เป้าหมายคือต้องการให้พิมพ์เลข `1, 2, 3, 4, 5` แต่มันดันพิมพ์ถึงแค่เลข `4`

```python
for i in range(1, 5):
    print(i)
    # ต้องแก้ตัวเลขในวงเล็บ range() อย่างไร?

```

**ปัญหาที่ 3: วาง `break` ผิดที่ (Misplaced Break)**
*ปัญหา:* โปรแกรมหยุดทำงานเร็วเกินไป มันควรจะค้นหาตัวอักษรต่อไป

```python
for char in "hello":
    if char == 'l':
        print("เจอตัว 'l' แล้ว หยุดการทำงาน!")
        break
    print(char)
    # ทำไมมันถึงพิมพ์แค่ 'h', 'e' แล้วจบเลยล่ะ?

```
