# กิจกรรมที่ 3.2: สร้างศิลปะบนหน้าจอด้วยลูปซ้อนลูป (Pattern Printing with Nested Loops)


---



**รูปแบบการทำกิจกรรม:** ความท้าทายเชิงความคิดสร้างสรรค์ (ทำเดี่ยว)

**จุดประสงค์:** เพื่อตอกย้ำความเข้าใจเรื่องการทำงานของลูปซ้อนลูป (Nested Loops) และแสดงให้เห็นว่าตัวแปรในลูปมีผลต่อการแสดงผลบนหน้าจออย่างไร ซึ่งช่วยฝึกกระบวนการคิดเชิงอัลกอริทึมและการแก้ปัญหา

**คำสั่งสำหรับนักศึกษา:**
ให้นักศึกษาเขียนโค้ด Python โดยใช้การวนลูปซ้อนลูป (Nested `for` loops) เพื่อพิมพ์ลวดลายศิลปะแบบ ASCII ตามโจทย์ที่กำหนดให้

**ตัวอย่างโจทย์พื้นฐาน:**

1. **สี่เหลี่ยมจัตุรัสด้วยดอกจัน:**

```text
    *****
    *****
    *****
    *****
    *****
```

2. **สามเหลี่ยมมุมฉากด้วยดอกจัน:**

```text
    *
    **
    ***
    ****
    *****
```



---
## Problem 1


---



````python
for i in range(5):
  for j in range(5):
    print("*", end="")
  print()
````



---
## Problem 2


---



````python
for i in range(5):
  for j in range(i+1):
    print("*", end="")
  print()
````
